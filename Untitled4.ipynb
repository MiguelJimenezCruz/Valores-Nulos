```python
data= pd.read_csv("Ventas_totales.csv")
data.head(5)
```

```python
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
data2=data.copy()
```

```python
data2["salon_ventas"]=data2["salon_ventas"].fillna(round(data["salon_ventas"].mean(),1))
#Se calcula el promedio de la columna y lo pone en los espacios vacios
```

```python
data2["tarjetas_debito"]=data2["tarjetas_debito"].fillna(round(data["salon_ventas"].mean(),1))
#Se calcula el promedio de la columna y lo pone en los espacios vacios
```

```python
data2["tarjetas_credito"]=data2["tarjetas_credito"].fillna(round(data["tarjetas_credito"].mean(),1))
#Se calcula el promedio de la columna y lo pone en los espacios vacios
```

```python
data2["otros_medios"]=data2["otros_medios"].fillna(method="ffill")
#Llena los espacios vacios con los datos de la siguiente casilla
```

```python
data2["bebidas"]=data2["bebidas"].fillna(method="ffill")
#Llena los espacios vacios con los datos de la siguiente casilla
```

```python
data2["subtotal_ventas_alimentos_bebidas"]=data2["subtotal_ventas_alimentos_bebidas"].fillna(round(data["subtotal_ventas_alimentos_bebidas"].median(),1))
#Llena las casillas vacias con el valor de la casilla de el medio de la columna
```

```python
data2["almacen"]=data2["almacen"].fillna(method="bfill")
#Llena los espacios vacios con los datos de la casilla anterior
```

```python
data2["panaderia"]=data2["panaderia"].fillna(method="bfill")
#Llena los espacios vacios con los datos de la casilla anterior
```

```python
data2["lacteos"]=data2["lacteos"].fillna(method="bfill")
#Llena los espacios vacios con los datos de la casilla anterior
```

```python
data2["carnes"]=data2["carnes"].fillna(method="bfill")
#Llena los espacios vacios con los datos de la casilla anterior
```

```python
data2["verduleria_fruteria"]=data2["verduleria_fruteria"].fillna(round(data["verduleria_fruteria"].median(),1))
#Llena las casillas vacias con el valor de la casilla de el medio de la columna
```

```python
data2["indumentaria_calzado_textiles_hogar"]=data2["indumentaria_calzado_textiles_hogar"].fillna(round(data["indumentaria_calzado_textiles_hogar"].median(),1))
#Llena las casillas vacias con el valor de la casilla de el medio de la columna
```

```python
data2["electronicos_articulos_hogar"]=data2["electronicos_articulos_hogar"].fillna(round(data["electronicos_articulos_hogar"].median(),1))
#Llena las casillas vacias con el valor de la casilla de el medio de la columna
```

```python
data2["alimentos_preparados_rotiseria"]=data2["alimentos_preparados_rotiseria"].fillna(round(data["alimentos_preparados_rotiseria"].median(),1))
#Llena las casillas vacias con el valor de la casilla de el medio de la columna
```

```python
data2["otros"]=data2["otros"].fillna(round(data["otros"].median(),1))
#Llena las casillas vacias con el valor de la casilla de el medio de la columna
```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```
